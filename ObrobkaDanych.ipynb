```python
import pandas as pd
```

```python
df = pd.read_csv('27__dane.csv', sep=';')

df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Imie"] = df["Name"].apply(lambda x: str(x).strip().split(" ")[0])
df["list Name"] = df["Name"].apply(lambda x: str(x).strip().split(" "))
uni = df["list Name"].apply(lambda x: len(x)).unique()
df['Nazwisko'] = df["Name"].apply(lambda x: str(x).strip().split(" ")[-1])
print(uni)
df


```

```python
df.drop(columns=['Name', "list Name"], inplace=True)
df
```

```python
df.rename(columns={"Age": "Wiek", "Address": "Adres", "Height": "Wzrost" , "Weight": 'Waga', }, inplace=True)
df
```

```python

df['BMI'] = df ['Waga']/ ((df['Wzrost']/ 100) ** 2)
df['BMI'] = df['BMI'].round(1)
df
```

```python
df= df.sort_values(ascending=False, by='BMI')
df
```

```python
df.to_csv('27__obrobione_dane.csv', sep=',', index=False)
```
